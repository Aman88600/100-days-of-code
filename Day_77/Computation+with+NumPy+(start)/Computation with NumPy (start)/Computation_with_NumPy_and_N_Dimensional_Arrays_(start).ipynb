#### **Challenge 4**: Print out all the indices of the non-zero elements in this array: [6,0,9,0,0,5,0]

```python
ch_4_array = [6,0,9,0,0,5,0]

# converting list to numpy array
ch_4_array = np.array(ch_4_array)

# using while loop
i = 0
while i < len(ch_4_array):
    if ch_4_array[i] != 0:
        print(f"{ch_4_array[i]} index = {i}")
    i += 1

# using numpy
indecies = np.nonzero(ch_4_array)
print(indecies)
```
